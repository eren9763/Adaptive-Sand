````python
# ── Cell 1: Mount Drive & configure paths ──────────────────────────────────
SEED = 42
import os, sys

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

ROOT_PATH = '/content/drive/MyDrive/SAND'
if ROOT_PATH not in sys.path:
    sys.path.insert(0, ROOT_PATH)

os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
os.environ['TF_CUDNN_DETERMINISTIC'] = '1'

print('Drive mounted, ROOT_PATH:', ROOT_PATH)
````

````python
# ── Cell 2: Copy datasets to Colab local storage (faster I/O) ──────────────
import shutil

drive_datasets = os.path.join(ROOT_PATH, 'datasets')
local_datasets = '/content/datasets'

if not os.path.exists(local_datasets):
    print('Copying datasets from Drive…')
    shutil.copytree(drive_datasets, local_datasets)
    print('Done.')
else:
    print('Datasets already present at', local_datasets)
````

````python
# ── Cell 3: Import Adaptive-SAND modules ───────────────────────────────────
from adaptive_sand import (
    SANDAdaptiveModel,
    SANDAdaptiveModelHybrid,
    SANDSmartPruningCallback,
    SANDSmartHybridPruningCallback,
    set_seeds, reset_seeds,
    dataset_to_numpy, load_prepared_dataset, build_classifier_mlp,
)
print('All Adaptive-SAND modules imported successfully.')
````

````python
# ── Cell 4: Run experiment ─────────────────────────────────────────────────
from run_experiment import run_all

DATASET   = 'mnist'   # 'mice' | 'isolet' | 'coil' | 'arcene' | 'har70'
EPOCHS    = 150
MODEL     = 'both'    # 'asand' | 'hybrid' | 'both' | 'all'
SEED      = 42

df = run_all(
    dataset_name=DATASET,
    epochs=EPOCHS,
    model_choice=MODEL,
    seed=SEED,
    results_dir=os.path.join(ROOT_PATH, 'results'),
)
````
